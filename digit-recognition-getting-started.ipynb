```python
train_data = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
train_data.head()
```

```python
test_data = pd.read_csv('/kaggle/input/digit-recognizer/test.csv')
test_data.head()
```

```python
# X, Y
from sklearn import preprocessing
x = train_data.drop(['label'], axis=1)
y = train_data['label']

mm_scaler = preprocessing.MinMaxScaler()
x = pd.DataFrame(mm_scaler.fit_transform(x))
#print(x)

#x = x.dropna(axis=1,how='all')
x = x.fillna(0)
print(x.shape)

from sklearn.decomposition import PCA
pca = PCA(30)
pca.fit(x)
x = pca.transform(x)
print(type(x))
print(x)

# Divide into train and validation set
x_train = x[0:40000,:]
print(x_train.shape)
y_train = y.loc[0:39999]
print(y_train.shape)
x_validate = x[40000:42000,:]
print(x_validate.shape)
y_validate = y.loc[40000:41999]
print(y_validate.shape)
```

```python
#classify with SVM
from sklearn import svm
from sklearn.metrics import accuracy_score, confusion_matrix

clf = svm.SVC(kernel='rbf', gamma='auto')
#fit the model to the train data
clf.fit(x_train, y_train)
#predict train data
pred_train = clf.predict(x_train)
#predict validate data
pred_validate = clf.predict(x_validate)
#print the results
train_accuracy_score = accuracy_score(y_train.to_numpy(), pred_train)
validate_accuracy_score = accuracy_score(y_validate.to_numpy(), pred_validate)
print(f"Train accuracy: {train_accuracy_score}, Validate accuracy: {validate_accuracy_score}")

#get the confusion matrix
print(confusion_matrix(y_validate.to_numpy(), pred_validate))
```
